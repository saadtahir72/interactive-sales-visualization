```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style for the plots
sns.set(style="whitegrid")
```

```python
#Visualize distributions of numerical features
plt.figure(figsize=(10, 6))
df['Total_Vehicle_Sales'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Vehicle Sales')
plt.xlabel('Total Vehicle Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# 1. Sales per Year (Line plot)
plt.figure(figsize=(10, 6))
sales_per_year = df.groupby('Year')['Total_Vehicle_Sales'].sum()
sns.lineplot(x=sales_per_year.index, y=sales_per_year.values, color='tab:blue', marker='o')
plt.title('Total Vehicle Sales per Year')
plt.xlabel('Year')
plt.ylabel('Total Vehicle Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Sales per Country (Bar plot)
plt.figure(figsize=(10, 6))
country_sales = df.groupby('Country')['Total_Vehicle_Sales'].sum().sort_values(ascending=False)
sns.barplot(x=country_sales.index, y=country_sales.values, palette='dark:#5A9_r')
plt.title('Total Vehicle Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Vehicle Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Create interactive line plot for sales per year
sales_per_year = df.groupby('Year')['Total_Vehicle_Sales'].sum().reset_index()

fig = px.line(sales_per_year, x='Year', y='Total_Vehicle_Sales',
              title='Total Vehicle Sales per Year',
              labels={'Total_Vehicle_Sales': 'Total Vehicle Sales', 'Year': 'Year'},
              markers=True)
fig.update_traces(line=dict(color='blue', width=2))
fig.show()
```

```python
# Create interactive bar plot for sales per country
country_sales = df.groupby('Country')['Total_Vehicle_Sales'].sum().reset_index()

fig = px.bar(country_sales, x='Country', y='Total_Vehicle_Sales',
             title='Total Vehicle Sales by Country',
             labels={'Total_Vehicle_Sales': 'Total Vehicle Sales', 'Country': 'Country'},
             color='Country', color_continuous_scale='Viridis')
fig.update_traces(marker=dict(line=dict(width=0)))
fig.show()
```

```python
# Create interactive stacked bar chart for car sales segments by age group
age_group_sales = df.groupby(['Customer_Age_Group', 'Market_Segment'])['Total_Vehicle_Sales'].sum().unstack().reset_index()

fig = px.bar(age_group_sales, x='Customer_Age_Group', y=age_group_sales.columns[1:],
             title='Car Sales Segments per Customer Age Group',
             labels={'Total_Vehicle_Sales': 'Total Vehicle Sales', 'Customer_Age_Group': 'Customer Age Group'},
             barmode='stack', color_discrete_sequence=px.colors.qualitative.Set3)
fig.show()
```

```python
# Create interactive heatmap for sales by region and market segment
region_segment_sales = df.groupby(['Region', 'Market_Segment'])['Total_Vehicle_Sales'].sum().unstack().reset_index()

fig = px.imshow(region_segment_sales.drop(columns='Region').values,
                labels={'x': 'Market Segment', 'y': 'Region', 'color': 'Total Vehicle Sales'},
                x=region_segment_sales.columns[1:], y=region_segment_sales['Region'],
                color_continuous_scale='Blues', title='Sales by Region and Market Segment')
fig.update_layout(xaxis={'title': 'Market Segment'}, yaxis={'title': 'Region'})
fig.show()
```

```python
import plotly.express as px

# Summarize total vehicle sales by country
country_sales = df.groupby('Country')['Total_Vehicle_Sales'].sum().reset_index()

# Use Plotly's built-in country-level choropleth map
fig = px.choropleth(country_sales,
                    locations='Country',  # Column with country names
                    locationmode='country names',  # Use country names for matching
                    color='Total_Vehicle_Sales',  # Column to define the color scale
                    color_continuous_scale='Viridis',  # Color scale
                    labels={'Total_Vehicle_Sales': 'Total Vehicle Sales'},  # Axis labels
                    title='Total Vehicle Sales by Country')

# Show the map
fig.show()
```
